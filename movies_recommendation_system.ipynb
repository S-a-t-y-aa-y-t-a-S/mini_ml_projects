import the dependencies

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

import the dataset

```python
from google.colab import drive
drive.mount("/content/drive")
```

```python
movies_data = pd.read_csv("/content/drive/MyDrive/self_projects/movies_reco/movies.csv")
movies_data.head()
```

```python
# checking the shape of the dataframe
movies_data.shape
```

```python
# checking for null values
movies_data.isnull().sum()
```

```python
# selected features
selected_features = ["genres","keywords","tagline","cast","director"]
```

```python
# replacing the null values of the selected features with empty string
for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna("")
```

```python
combined_feature = movies_data["genres"]+" "+movies_data["keywords"]+" "+movies_data["tagline"]+" "+movies_data["cast"]+" "+movies_data["director"]
```

```python
print(combined_feature)
```

Converting the text data into numerical data

```python
vectorizer = TfidfVectorizer()
```

```python
vectorized_features = vectorizer.fit_transform(combined_feature)
```

```python
vectorized_features.toarray()
```

```python
similarity = cosine_similarity(vectorized_features)
similarity
```

```python
similarity.shape
```

```python
# getting the movie name from the user

movie_name = input("Enter the movie name: ")
```

```python
# creating a list of all movie name

list_of_all_titles = movies_data["title"].tolist()
print(list_of_all_titles)
```

```python
# finding close match with respect to the movie name stated

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# finding the index of the movie with title

movie_index = movies_data[movies_data["title"] == close_match]["index"].values[0]
movie_index
```

```python
# getting a list of similar movies

similarity_score = list(enumerate(similarity[movie_index]))
similarity_score
```

```python
# sorting the movies based on the similarity score (list of tuples)
# sorting on the basis of the 2nd element of each tuple

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse=True)
sorted_similar_movies
```

```python
len(sorted_similar_movies)
```

```python
i = 0

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data["index"] == index]["title"].values[0]

  if i < 30:
    print(i+1, " : ", title_from_index)
    i += 1
```

Movie recommendation system

```python
input_movie_name = input("Enter your favourite movie name: ")

movie_list = movies_data["title"].tolist()

close_match_list = difflib.get_close_matches(input_movie_name, movie_list)

close_match = close_match_list[0]

index = movies_data[movies_data["title"] == close_match]["index"].values[0]

similarity_score = list(enumerate(similarity[index]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse=True)

print("Movies suggested for you:\n")

i = 0

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data["index"] == index]["title"].values[0]

  if i < 30:
    print(i+1, " : ", title_from_index)
    i += 1
```
