importing the dependencies

```python
import numpy as np
import pandas as pd
import sklearn.datasets

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

importing the dataset

```python
imported_dataset = sklearn.datasets.load_breast_cancer()
breast_data = pd.DataFrame(data = imported_dataset.data, columns = imported_dataset.feature_names)
```

```python
# putting the labelled data as well
breast_data["label"] = imported_dataset.target
```

```python
# checking the shape of the dataframe
breast_data.shape
```

```python
# checking for the first five rows
breast_data.head()
```

```python
# checking for the last 5 rows
breast_data.tail()
```

```python
# checking for any null value
breast_data.isnull().sum()
```

```python
# checking the distribution of the labelled data
breast_data["label"].value_counts()
```

- 1 -> Benign
- 0 -> Malignant

```python
# displaying the statistical features of each column
breast_data.describe()
```

```python
# displaying other information
breast_data.info()
```

```python
# finding the mean value of each column based on the label data
breast_data.groupby("label").mean()
```

```python
# splitting the data into features and target
X = breast_data.drop(columns = "label", axis=1)
y = breast_data["label"]
```

```python
# splitting the data into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=2)
```

```python
X.shape, X_train.shape, X_test.shape
```

```python
y.shape, y_train.shape, y_test.shape
```

training the model

```python
lor_model = LogisticRegression(max_iter=5000)
```

```python
lor_model.fit(X_train, y_train)
```

```python
y_pred_train = lor_model.predict(X_train)
y_pred_test = lor_model.predict(X_test)
```

accuracy score

```python
# training data
print("The accuracy score of the training data: ", accuracy_score(y_train, y_pred_train))

# testing data
print("The accuracy score of the testing data: ", accuracy_score(y_test, y_pred_test))
```
